#Pre_Trained Models

```python
# Text Classification
import transformers
from transformers import pipeline
classifier = pipeline('text-classification', model='roberta-large-mnli')
```

```python
Task = "The Patient is About to die today evening"
classifier(Task)
```

```python
task = "Hai The Movie is really very good I suggest everyone to spend some time for this movie"
classifier(task)
```

```python
# Zero-Shot Classification


from transformers import pipeline
classifier = pipeline('zero-shot-classification', model='roberta-large-mnli')
```

```python


sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(sequence_to_classify, candidate_labels)
```

```python
# Token-classification (NER (Named Entity Recognition))

from transformers import pipeline
classifier = pipeline('token-classification', model='cahya/bert-base-indonesian-NER')

```

```python
Task = 'Kamal want to Travel to USA to attend the Interview and get back to India via Flight'
classifier(Task)
```

```python
# Fill Mask

from transformers import pipeline
classifier = pipeline('fill-mask', model='google-bert/bert-base-uncased')

```

```python
task = 'I am having interest in [MASK] intelligence'
classifier(task)
```
