```python
import numpy as np
import transformers
```

```python
from transformers import pipeline
```

#MLM (Mask Language Modelling)

```python
unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("The man worked as a [MASK].")
```

```python
unmasker = pipeline('text-classification', model='bert-base-uncased')
unmasker("The Movie is very good to watch")
```

```python
unmasker = pipeline('token-classification', model='bert-base-uncased')
unmasker("The Movie is very good to watch")
```
